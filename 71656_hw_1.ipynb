After one hot encoding the numeric data and combining it with the one hot encoded  
categorical data I discovered that the algorithm is overfitting.  
So I experimented a bit and discovered that removing the following columns gives the best result

```python
train = train.drop(['PoolArea', 'MiscVal', 'GarageFinish','FireplaceQu', 'LotConfig', 'LandContour', 'LandSlope' ], axis=1)
```

```python
train = train.drop(['Street', 'Utilities', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)
```

```python
categorical_column_names = train.dtypes[train.dtypes == 'object'].index
print(categorical_column_names)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target = train.SalePrice
print(train.shape)
train = train.drop(['SalePrice'], axis=1)
print(train.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_categorical_column_names = ['MSSubClass', 'OverallQual', 'OverallCond', 'MoSold', 'YrSold', 'Fireplaces', 'BedroomAbvGr' ]
numeric_column_names = train.describe().columns
print(numeric_categorical_column_names)
print(numeric_column_names)
```

```python
def get_categorical_numeric_columns(data, column_names):
    return pd.DataFrame({col: data[col] for col in column_names}, 
                        index = data.index)

numeric_categorical_columns = get_categorical_numeric_columns(train, numeric_categorical_column_names)
numeric_categorical_columns
one_hot_encoder = OneHotEncoder().fit(numeric_categorical_columns)
one_hot_encoded_numeric_categorical = one_hot_encoder.transform(numeric_categorical_columns)
one_hot_encoded_numeric_categorical.todense()
```

```python
def delete_and_concatenate_one_hot_encoded_columns(data, encoded_numeric_columns, column_names):
    data = data.fillna(0)
    data = data.drop(column_names, axis=1)
    return pd.np.concatenate([encoded_numeric_columns.todense(), data], axis=1)
one_hot_encoded_numeric_columns = delete_and_concatenate_one_hot_encoded_columns(train[numeric_column_names], one_hot_encoded_numeric_categorical, numeric_categorical_column_names)
one_hot_encoded_numeric_columns.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
numeric_x_train, numeric_x_test, numeric_y_train, numeric_y_test = train_test_split(one_hot_encoded_numeric_columns, target, test_size=0.3, random_state=4330)
```

```python
regressor = LinearRegression().fit(numeric_x_train, numeric_y_train)
print("Train score: {}".format(regressor.score(numeric_x_train, numeric_y_train)))
print("Test score: {}".format(regressor.score(numeric_x_test, numeric_y_test)))
```

```python
train = train.fillna("")
encoders = {col: LabelEncoder().fit(train[col]) for col in categorical_column_names}
encoders
```

```python
def encode_categorical_data(data, columns, encoders):
    return pd.DataFrame({col: encoders[col].transform(data[col]) for col in columns},
                        index = data.index)

train_encoded = encode_categorical_data(train, categorical_column_names, encoders)
train_encoded
```

```python
one_hot_encoder = OneHotEncoder().fit(train_encoded)
one_hot_train = one_hot_encoder.transform(train_encoded)
one_hot_train[:8].todense()
```

```python
categorical_x_train, categorical_x_test, categorical_y_train, categorical_y_test = train_test_split(one_hot_train, target, test_size=0.3, random_state=4330)
```

```python
cat_regression = LinearRegression().fit(categorical_x_train, categorical_y_train)
print("Train score: {}".format(cat_regression.score(categorical_x_train, categorical_y_train)))
print("Test score: {}".format(cat_regression.score(categorical_x_test, categorical_y_test)))
```

```python
new_x_train = pd.np.concatenate([categorical_x_train.todense(), numeric_x_train], axis=1)
new_x_test = pd.np.concatenate([categorical_x_test.todense(), numeric_x_test], axis=1)
print(new_x_train.shape, new_x_test.shape)
```

```python
all_data_lr = LinearRegression().fit(new_x_train, numeric_y_train)
print(all_data_lr.score(new_x_train, numeric_y_train))
print(all_data_lr.score(new_x_test, numeric_y_test))
```

The algorithm seems to be overfitting. We need to add Regularisation


```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

alpha = [0.01, 0.1, 1, 10, 100, 110, 120]

for a in alpha:
    all_data_lr = Ridge(alpha = a).fit(new_x_train, numeric_y_train)
    print('alpha:', a)
    print(all_data_lr.score(new_x_train, numeric_y_train), all_data_lr.score(new_x_test, numeric_y_test))
    print()
```

```python
alpha = [0.01, 0.1, 1, 10, 100, 110, 120]

for a in alpha:
    all_data_lr = Lasso(alpha = a).fit(new_x_train, numeric_y_train)
    print('alpha:', a)
    print(all_data_lr.score(new_x_train, numeric_y_train), all_data_lr.score(new_x_test, numeric_y_test))
    print()
```

```python
alphas = [0.01, 0.1, 1, 10, 100, 1000]
for alpha in alphas:
    all_data_lr = Ridge(alpha = alpha).fit(new_x_train, pd.np.log10(numeric_y_train))
    print('alpha:', alpha)
    print(all_data_lr.score(new_x_train, pd.np.log10(numeric_y_train)), all_data_lr.score(new_x_test, pd.np.log10(numeric_y_test)))
    print()
```

```python
alphas = [0.01, 0.1, 1, 10, 100, 1000]
for alpha in alphas:
    all_data_lr = Lasso(alpha = alpha).fit(new_x_train, pd.np.log10(numeric_y_train))
    print('alpha:', alpha)
    print(all_data_lr.score(new_x_train, pd.np.log10(numeric_y_train)), all_data_lr.score(new_x_test, pd.np.log10(numeric_y_test)))
    print()
```

Not a good idea

```python
from sklearn.metrics import r2_score

for alpha in [0.0001, .001,0.01, 0.1, 1, 10, 100]:
    all_data_lr = Ridge(alpha = alpha).fit(new_x_train, pd.np.log10(numeric_y_train))
    print('alpha:', alpha)
    print(r2_score(10**all_data_lr.predict(new_x_train), numeric_y_train))
    print(r2_score(10**all_data_lr.predict(new_x_test), numeric_y_test))
    print()
```
